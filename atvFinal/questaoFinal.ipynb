```python
import pandas as pd
import json


with open('questaoFinal.json', 'r') as file:
    data = json.load(file)  

df = pd.json_normalize(data)
df
```

```python
df_detalhes = pd.json_normalize(data, meta=['id_loja', 'nome_loja'], record_path=['detalhes'])
df_vendas = pd.json_normalize(data, record_path=['vendas'])
df = pd.concat([df_detalhes, df_vendas], axis=1)
df
```

```python
#também é possivel dessa forma:
import pandas as pd
import json

with open('questaoFinal.json', 'r') as file:
    dados = json.load(file)  

# 1. Normalizando as listas, incluindo os dados de nível superior
df_details = pd.json_normalize(dados, record_path='detalhes', meta=['id_loja', 'nome_loja'])
df_vendas = pd.json_normalize(dados, record_path='vendas', meta=['id_loja', 'nome_loja'])

# 2. Combinando os DataFrames
df = pd.merge(df_details, df_vendas, on=['id_loja', 'nome_loja'])
df
```
